```python
import numpy as np
import pandas as pd
from numpy.linalg import norm, det, inv, eig
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from scipy.linalg import lu, svd
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('student_data_200-2.csv',index_col=0)
df
X = df.values
print(X)
```

```python

X_numeric = df.select_dtypes(include=[np.number])

X = X_numeric.values

v1 = X[0]

norm_1 = norm(v1, 1)   # L1 norm
norm_2 = norm(v1, 2)   # L2 norm

print("v1 =", v1)
print("L1 norm =", norm_1)
print("L2 norm =", norm_2)
```

```python
X = df.select_dtypes(include=[np.number]).values

v1 = X[0]
v2 = X[1]

dot_product = np.dot(v1, v2)

angle = np.arccos(
    dot_product / (norm(v1) * norm(v2))
) * (180 / np.pi)

print("Dot product:", dot_product)
print("Angle (degrees):", angle)
```

```python
A = X[:5]
B = X[5:10]

matrix_addition = A + B
matrix_multiplication = np.dot(A, B.T)
matrix_addition, matrix_multiplication
```

```python
cov_matrix = np.cov(X.T)
eigenvalues, eigenvectors = eig(cov_matrix)

eigenvalues, eigenvectors
```
